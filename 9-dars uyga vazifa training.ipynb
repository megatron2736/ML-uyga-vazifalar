# missing valuelarni tuldirish

```python
df.isnull().sum()
```

```python
bush_katak=df.isnull().sum()>0
```

```python
bush_katak
```

```python
for col in df.columns[bush_katak]:
    if df[col].dtype=='object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)
```
